```python
from pyspark.sql.functions import expr
from pyspark.sql.functions import col, sum
```

```python
bronze_sales_path = "/Volumes/workspace/m5_project/bronze/bronze_sales"

bronze_sales_df = spark.read.parquet(
    bronze_sales_path
)

display(bronze_sales_df.limit(5))
```

```python
day_columns = [col for col in bronze_sales_df.columns if col.startswith("d_")]

print("Total day columns:", len(day_columns))
print(day_columns[:10])
```

```python
display(
    bronze_sales_df.select(
        "id",
        "item_id",
        "dept_id",
        "cat_id",
        "store_id",
        "state_id"
    ).limit(10)
)
```

```python
id_columns = [
    "id",
    "item_id",
    "dept_id",
    "cat_id",
    "store_id",
    "state_id"
]
```

```python

#Unpivot
n = len(day_columns)

stack_expr = ", ".join(
    [f"'{col}', {col}" for col in day_columns]
)

silver_sales_df = bronze_sales_df.selectExpr(
    *id_columns,
    f"stack({n}, {stack_expr}) as (day, sales)"
)

display(silver_sales_df.limit(20))
```

```python
print("Original Bronze Rows:", bronze_sales_df.count())
print("Silver Rows After Unpivot:", silver_sales_df.count())
```

```python
#check for null values

silver_sales_df.select([
    sum(col(c).isNull().cast("int")).alias(c)
    for c in silver_sales_df.columns
]).display()
```

```python
#Duplicate Records

duplicate_check = silver_sales_df.groupBy(
    "id", "day"
).count().filter("count > 1")

display(duplicate_check)
```

```python
#Negative Sales

negative_sales = silver_sales_df.filter(
    col("sales") < 0
)

display(negative_sales)
```

```python
#Distinct Days

print(
    silver_sales_df.select("day")
    .distinct()
    .count()
)
```

```python
calendar_df = spark.read.parquet(
    "/Volumes/workspace/m5_project/bronze/bronze_calendar"
)
```

```python

silver_sales_df = silver_sales_df.join(
    calendar_df.select("d", "date", "wm_yr_wk"),
    silver_sales_df.day == calendar_df.d,
    "left"
).drop("d")

display(silver_sales_df.limit(10))
```

```python
missing_dates = silver_sales_df.filter(
    col("date").isNull()
)

display(missing_dates)
```

```python
silver_sales_path = "/Volumes/workspace/m5_project/silver/silver_sales"

silver_sales_df.write.mode("overwrite").parquet(
    silver_sales_path
)

print("Silver Sales written successfully")
```

```python
silver_check_df = spark.read.parquet(
    silver_sales_path
)

display(silver_check_df.limit(10))
```
